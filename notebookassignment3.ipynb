```python
train_inputs = np.load("train_inputs.npy")  # shape: [N, 32, 64]

# Visualize the first 6 inputs
num_samples = 6
plt.figure(figsize=(15, 6))

for i in range(num_samples):
    plt.subplot(2, 3, i + 1)
    plt.imshow(train_inputs[i], cmap='gray', interpolation='nearest')
    plt.title(f"Sample {i}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
